```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D

x = [4,6,8,10] # Order number
y1 = [471,652,828,1000] # Non-trans w/o
y2 = [292,395,500,598]
y3 = [253,361,423,482]
y4 = [209,231,243,258]

coef1 = np.polyfit(x,y1,1)
poly1d_fn = np.poly1d(coef1)

coef2 = np.polyfit(x,y2,1)
poly1d_fn1 = np.poly1d(coef2)

coef3 = np.polyfit(x,y3,1)
poly1d_fn2 = np.poly1d(coef3)

coef4 = np.polyfit(x,y4,1)
poly1d_fn3 = np.poly1d(coef4)

# poly1d_fn is now a function which takes in x and returns an estimate for y

plt.plot(x,y1, 'yo') 
plt.plot(x, poly1d_fn(x), '--k', label = 'Non-Transposed w/o')
plt.plot(x,y2, 'yo') 
plt.plot(x, poly1d_fn1(x), '--b',label = 'Transposed w/o')
plt.plot(x,y3, 'yo') 
plt.plot(x, poly1d_fn2(x), '--r',label = 'Non-Transposed o2')
plt.plot(x,y4, 'yo') 
plt.plot(x, poly1d_fn3(x), '--p',label = 'Transposed o2')
plt.title('Comparison of Performance')
plt.legend(loc='upper left', borderaxespad=0.8)
plt.grid()
```
